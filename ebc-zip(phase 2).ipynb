```python
import os
import numpy as np

dir1 = "/home/gpuuser6/mahek/Semi-Supervised-Crowd-Counting-BTech-Thesis/ZIP/labels"
dir2 = "/home/gpuuser6/mahek/Semi-Supervised-Crowd-Counting-BTech-Thesis/ZIP/data/sha/train/labels"

files1 = sorted([f for f in os.listdir(dir1) if f.endswith(".npy")])
files2 = sorted([f for f in os.listdir(dir2) if f.endswith(".npy")])

# -------------------------------
# Step 1: Check file lists
# -------------------------------
if files1 != files2:
    print("❌ File lists differ!")

    only1 = set(files1) - set(files2)
    only2 = set(files2) - set(files1)

    if only1:
        print("Only in dir1:", only1)
    if only2:
        print("Only in dir2:", only2)

else:
    print("✅ Same file names\n")

    # -------------------------------
    # Step 2: Compare contents
    # -------------------------------
    all_same = True

    for f in files1:
        path1 = os.path.join(dir1, f)
        path2 = os.path.join(dir2, f)

        a = np.load(path1)
        b = np.load(path2)

        # Shape check (fixes your error)
        if a.shape != b.shape:
            print(f"❌ Shape mismatch in {f}: {a.shape} vs {b.shape}")
            all_same = False
            continue

        # Dtype check
        if a.dtype != b.dtype:
            print(f"⚠️ Dtype mismatch in {f}: {a.dtype} vs {b.dtype}")

        # Value check
        if not np.allclose(a, b, atol=1e-6):
            diff = np.abs(a - b)
            print(f"❌ Value mismatch in {f}")
            print(f"   Max diff: {diff.max()}")
            print(f"   Mean diff: {diff.mean()}")

            # Optional: count mismatch (useful for crowd counting)
            print(f"   Sum1: {a.sum()}, Sum2: {b.sum()}")

            all_same = False

    # -------------------------------
    # Final result
    # -------------------------------
    if all_same:
        print("🎉 All files are identical (within tolerance)")
    else:
        print("\n⚠️ Directories are NOT identical")
```

```python
import scipy.io as sio
import numpy as np
import os

mat_dir = '/home/gpuuser6/mahek/Dataset/ShanghaiTech/part_A/train_data/ground-truth'   # your .mat files
out_dir = '/home/gpuuser6/mahek/Semi-Supervised-Crowd-Counting-BTech-Thesis/ZIP/data/sha/train/labels/'
os.makedirs(out_dir, exist_ok=True)

mat_files = sorted([f for f in os.listdir(mat_dir) if f.endswith('.mat')])

for idx, mat_file in enumerate(mat_files, start=1):
    mat = sio.loadmat(os.path.join(mat_dir, mat_file))
    points = mat['image_info'][0][0][0][0][0]  # (N, 2) array of (x, y)
    
    out_name = f"{idx:03d}.npy"  # → 001.npy, 002.npy, ...
    np.save(os.path.join(out_dir, out_name), points.astype(np.float32))
    print(f"{mat_file} → {out_name}, {points.shape[0]} people")
```
